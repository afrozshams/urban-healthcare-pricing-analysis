```python
import pandas as pd
```

```python
df = pd.read_csv("Project_DataSets02.csv")
df.head()
```

```python
df.info()
```

```python
metro = ["Hyderabad", "Bangalore", "Chennai", "Kolkata", "Mumbai", "Delhi"]
city_type = []

for city in df["City"]:
    
    if city in metro:
        city_type.append("Metro")
        
    else:
        city_type.append("Non-Metro")
```

```python
df.insert(df.columns.get_loc("City")+1, "City_Tier", city_type)
```
